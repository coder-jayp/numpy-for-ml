````python
# Two clusters
np.random.seed(42)
class0 = np.random.randn(50, 2) + np.array([-2, -2])
class1 = np.random.randn(50, 2) + np.array([2, 2])

plt.scatter(class0[:,0], class0[:,1], label="Class 0")
plt.scatter(class1[:,0], class1[:,1], label="Class 1")
plt.legend()
plt.title("Synthetic 2D Dataset")
plt.show()
````

````python
# Dataset
X = np.vstack([class0, class1])
y = np.array([0]*50 + [1]*50)

# Add bias term
X_b = np.c_[np.ones((X.shape[0], 1)), X]

# Sigmoid function
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Initialize weights
theta = np.random.randn(X_b.shape[1])

# Training loop
lr = 0.1
losses = []
for epoch in range(1000):
    z = X_b @ theta
    h = sigmoid(z)
    gradient = X_b.T @ (h - y) / len(y)
    theta -= lr * gradient
    # Binary cross-entropy loss
    loss = -np.mean(y*np.log(h+1e-8) + (1-y)*np.log(1-h+1e-8))
    losses.append(loss)

plt.plot(losses)
plt.title("Training Loss Curve")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.show()
````

````python
# Grid for decision boundary
aa, bb = np.meshgrid(np.linspace(-5, 5, 200), np.linspace(-5, 5, 200))  # finer grid
grid = np.c_[np.ones((aa.size, 1)), aa.ravel(), bb.ravel()]
probs = sigmoid(grid @ theta).reshape(aa.shape)

# Plot decision boundary
plt.figure(figsize=(7,5))
plt.contourf(aa, bb, probs, levels=[0, 0.5, 1], alpha=0.3, colors=["blue", "red"])  # light overlay
plt.scatter(class0[:, 0], class0[:, 1], c='blue', edgecolor='k', label="Class 0")
plt.scatter(class1[:, 0], class1[:, 1], c='red', edgecolor='k', label="Class 1")
plt.legend()
plt.title("Logistic Regression Decision Boundary")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
````
